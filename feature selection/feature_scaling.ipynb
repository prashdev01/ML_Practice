# Standerdazing

```python
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill =list(df['total_bill'])
sns.displot(total_bill)
```

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
normalized_data = []
for i in total_bill:
    z_score = (i-mean)/std
    normalized_data.append(z_score)
    
normalized_data
```

```python
sns.displot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd


pd.DataFrame(scaler.transform(df[['total_bill','tip']]))
```
