```python
df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']
```

```python
df['avarage_score'] = df['total_score']/3
```

```python
df.head()
```

```python
# exploring more visulizstion
fig,axis=plt.subplots(1,2,figsize =(15,7))
plt.subplot(121)
sns.histplot(data=df,x='avarage_score',bins=30,kde=True, color='g')
plt.subplot(122)
sns.histplot(data=df,x='avarage_score',bins=30,kde=True,hue='gender')
```

```python
fig,axis=plt.subplots(1,3,figsize =(20,7))
plt.subplot(141)
sns.histplot(data=df,x='avarage_score',bins=30,kde=True, hue='lunch')
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='avarage_score',bins=30,kde=True,hue='lunch')
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],x='avarage_score',bins=30,kde=True, hue='lunch')
```
